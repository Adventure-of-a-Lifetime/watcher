```python
import logging
import pandas as pd
import json

logging.basicConfig(level=logging.INFO,
                    format="%(asctime)s %(levelname)s %(message)s")
logging.info("inited")
pd.set_option('max_colwidth',50)  
pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', None)
```

```python
def display_domain(path):
    dfs = []
    with open(path, "r") as f:
        for line in f.readlines():
            line = line.strip('\n')
            if line.split() and line.find('no_results') == -1:
                json_data = pd.json_normalize(json.loads(line),'urls',['domain','blacklists'])
                json_data = json_data[['domain','url','tags','blacklists']]
                dfs.append(json_data)
                
    df = pd.concat(dfs, sort=False)
    print(df)
    return df
```

```python
display_domain('E:/Project/Python/ngrok_domains_result.json')
```
